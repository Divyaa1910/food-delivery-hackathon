```python
import pandas as pd
orders = pd.read_csv("orders.csv")
orders.head()
```

```python
users = pd.read_json("users.json")
users.head()
```

```python
import sqlite3
conn = sqlite3.connect(':memory:')
with open('restaurants.sql', 'r') as f:
    conn.executescript(f.read())
restaurants = pd.read_sql_query("SELECT * FROM restaurants", conn)
conn.close()
restaurants.head()
```

```python
merged1 = orders.merge(users, on='user_id', how='left')
merged1.head()
```

```python
final_data = merged1.merge(restaurants, on='restaurant_id', how='left')
final_data.head()
```

```python
final_data.to_csv("final_merged.csv", index=False)
print("File saved!")
```
